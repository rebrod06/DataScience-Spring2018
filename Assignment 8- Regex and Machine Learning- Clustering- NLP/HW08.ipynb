### Task 1.2 k-means cluster analysis
1. Scale the dataset using the *scale* function of the sklearn.preprocessing library. 
+ Using k-means, cluster the states into four clusters. Which states belong to which clusters?
+ Vary k and find the *best* value. How do you determine *best*? 

````python
#Task 1.2.1: Scale dataset

crime_types = list(crime)
crime_types.remove('State')
scaled_crime = scale(crime[crime_types])

scaled_crime[0:5,:]
````

````python
#Task 1.2.2: Use k-means to cluster into 4 clusers

X = scaled_crime
model = KMeans(n_clusters = 4)
y_crime_pred_kmean = model.fit_predict(X)

states = crime['State']
states = states.values

for i in range(0,4):
    states_in_cluster = states[np.where(y_crime_pred_kmean==i)]
    print("States in cluster ", i+1, ": ", states_in_cluster, "\n")
````

````python
#Task 1.2.3: Find best k

#test different k's to find best score
ks = range(1,20)
scores = []
for k in ks:
    model = KMeans(n_clusters=k)
    model.fit_predict(X)
    scores.append(-model.score(X))
    
plt.plot(ks, scores)
plt.xlabel('k')
plt.ylabel('total inntra-cluster distance')
plt.title("K-Means Scores for Various k's")
plt.xticks(ks)
plt.show()
````

**Interpretation:** 

Our k-means score with k=4 clusters was 57.

To determine the "best" k for our clustering, we want to pick the k where total intra-cluster distance tapers off. Looking at our plot above, we can see the most prominent "elbow" (i.e. the point of tapering) is at k=4. Thus, our best choice for k is 4.

### Task 1.3 Hierarchical cluster analysis

````python
#Task 1.3: Hierarchical clustering

#hierarchical clustering w/ complete linkage and Euclidean dist, n=4
agg_cluster_model = AgglomerativeClustering(linkage='complete', affinity='euclidean', n_clusters=4)
y_crime_pred_agg = agg_cluster_model.fit_predict(X)

for i in range(0,4):
    st_in_cluster_agg = states[np.where(y_crime_pred_agg==i)]
    st_in_cluster_kmean = states[np.where(y_crime_pred_kmean==i)]
    print('States in hierarchical cluster ', i+1, ": ", st_in_cluster_agg) 
````

````python
#compare clusters from k-means and hierarchical
dict = {'States':states, 'k_cluster': y_crime_pred_kmean, 'h_cluster': y_crime_pred_agg}
comp_df = pd.DataFrame(data = dict)
comp_df = comp_df[['States', 'k_cluster', 'h_cluster']]
comp_df['Matching'] = np.where(comp_df['k_cluster']-comp_df['h_cluster']==0, 'match', 'mismatch')
print(comp_df)
````

````python
text = """To learn about pros/cons of data science, go to http://datascience.net.\
Alternatively, go to datascience.net/2018/"""
````

````python
#Task 2.1: extract urls

re.findall(r"http://\w*\.\w*|\w+\.\w*", text)
# http:// --> finds exact match of 'http://'
# \w*\.\w* --> finds 0 or more words, a period, 0 or more words
# | --> logical 'or'
# \w+ --> finds 1 or more words
````


Alpha is also known as the smoothing factor and ranges from 0 (no smoothing) to 1 (Laplace Smoothing). You can experiment with different values to see if you get better results. 


````python
#Task 3.2a: Vectorize Data

#create training and test data sets
corpus_train = fetch_20newsgroups(subset='train')
corpus_test = fetch_20newsgroups(subset='test')

#vectorize training and test data using TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_train = tfidf_vectorizer.fit_transform(corpus_train.data)
tfidf_test = tfidf_vectorizer.transform(corpus_test.data)

#vectorize training and test data using CountVectorizer
counter_vectorizer = CountVectorizer()
counter_train = counter_vectorizer.fit_transform(corpus_train.data)
counter_test = counter_vectorizer.transform(corpus_test.data)

# vectorize using built-in function
#vectorized_corpus = fetch_20newsgroups_vectorized(subset='all')
````

````python
#inspect vectorized datasets

print(tfidf_train.shape)
print(tfidf_test.shape)

print(counter_train.shape)
print(counter_test.shape)

print(corpus_test.target.shape)
````

````python
#Task 3.2b: MNB Classifier for tfidf-vectorized data

#instantiate classifier
mnb_clf = MultinomialNB(alpha = 0.1)

#fit (i.e. train) model to tfidf-vectorized data 
# model_tfidt = mnb_clf.fit(tfidf_train, corpus_train.target)
mnb_clf.fit(tfidf_train, corpus_train.target)

#use model on test data
predict_tfidf = mnb_clf.predict(tfidf_test)

#calc accuracy and F-score
print("Tfidf-vectorized Data with MNB alpha=.01")
#print("Accuracy: ", mnb_clf.score(tfidf_test, corpus_test.target))
print("Accuracy: ", metrics.accuracy_score(corpus_test.target, predict_tfidf))
print("F-score: ", metrics.f1_score(corpus_test.target, predict_tfidf, average='macro'))
````

````python
#Task 3.2b Cont: MNB Classifier for counter-vectorized data

#instantiate classifier
mnb_clf = MultinomialNB(alpha = 0.1)

#fit (i.e. train) model to counter-vectorized data
mnb_clf.fit(counter_train, corpus_train.target)

#use model on test data
predict_counter = mnb_clf.predict(counter_test)

#calc accuracy and F-score
print("Counter-vectorized Data with MNB alpha=.01")
print("Accuracy: ", metrics.accuracy_score(corpus_test.target, predict_counter))
print("F-score: ", metrics.f1_score(corpus_test.target, predict_counter, average='macro'))
````

````python
#Task 3.2b Cont: experiment w/ different alphas for tfidf-vectorized data

#alphas must be in range (0,1]
alphas = list(np.linspace(0.01, 1, num=50, endpoint=True))

#calc accuracy of MNB model using different alphas
tfidf_scores = []
for i in alphas:
    mnb_clf = MultinomialNB(alpha = i)
    mnb_clf.fit(tfidf_train, corpus_train.target)
    predict_tfidf = mnb_clf.predict(tfidf_test)
    tfidf_scores.append(metrics.accuracy_score(corpus_test.target, predict_tfidf))
    
#print(tfidf_scores)
````

````python
#Task 3.2b Cont: experiment w/ different alphas for counter-vectorized data

#calc accuracy using different alphas
counter_scores = []
for i in alphas:
    mnb_clf = MultinomialNB(alpha = i)
    mnb_clf.fit(counter_train, corpus_train.target)
    predict_counter = mnb_clf.predict(counter_test)
    counter_scores.append(metrics.accuracy_score(corpus_test.target, predict_counter))
    
#print(counter_scores)
````

````python
#plot alphas and model accuracies

plt.plot(alphas, tfidf_scores, label="tfidf-vectorized")
plt.plot(alphas, counter_scores, label="counter-vectorized")
plt.xlabel("Alpha")
plt.ylabel("Accuracy")
plt.title("Mean Accuracy of Multinomial Naive Bayes Classifier \n with alpha values from .01 to 1")
plt.legend()
````

We can see that as we increase alpha, in general, the accuracy of the MNB classifier on our test dataset decreases. Very small alpha (closest to 0) provides the best accuracy when the data has been tfidf-vectorized, whereas accuracy for counter-vectorized data remains fairly constant when alpha is between 0 and 0.5. See below for the optimized alpha value for each.

````python
#find best accuracy and corresponding alpha for each

best_alpha_tfidf = alphas[tfidf_scores.index(max(tfidf_scores))]
best_accuracy_tfidf = max(tfidf_scores)
print("Tfidf_vectorized data")
print("Best alpha: ", best_alpha_tfidf)
print("Accuracy: ", best_accuracy_tfidf, "\n")

best_alpha_counter = alphas[counter_scores.index(max(counter_scores))]
best_accuracy_counter = max(counter_scores)
print("Counter_vectorized data")
print("Best alpha: ", best_alpha_counter)
print("Accuracy: ", best_accuracy_counter)
````

#### Removing Stopwords

Now we'll use the NLTK stopword list to improve our data vectors. TfidfVectorizer and CountVectorizer both can take an argument called stop_words. The words passed to this arguement are considered as stopwords and are not vectorized. Use the stopwords list from NLTK and pass it to vectorizers. Then evalulate the new vectors using Multinomial Naive Bayes.

````python
#Task 3.2c: removing stop words

#stop words
sw = nltk.corpus.stopwords.words('english')

#vectorize training and test data w/out stopwords using TfidfVectorizer
tfidf_vectorizer_sw = TfidfVectorizer(stop_words=sw)
tfidf_train_sw = tfidf_vectorizer_sw.fit_transform(corpus_train.data)
tfidf_test_sw = tfidf_vectorizer_sw.transform(corpus_test.data)

#vectorize training and test data w/out stopwords using CountVectorizer
counter_vectorizer_sw = CountVectorizer(stop_words=sw)
counter_train_sw = counter_vectorizer_sw.fit_transform(corpus_train.data)
counter_test_sw = counter_vectorizer_sw.transform(corpus_test.data)
````

````python
#Task 3.2c cont: MNB Classifier for tfidf-vectorized data (no stopwords)

#instantiate classifier
mnb_clf = MultinomialNB(alpha = best_alpha_tfidf)

#fit (i.e. train) model to tfidf-vectorized data 
mnb_clf.fit(tfidf_train_sw, corpus_train.target)

#use model on test data
predict_tfidf_sw = mnb_clf.predict(tfidf_test_sw)

#calc accuracy and F-score
accuracy_tfidf_sw = metrics.accuracy_score(corpus_test.target, predict_tfidf_sw)
print("Tfidf-vectorized Data (no stop words) with MNB and best alpha")
print("Accuracy: ", accuracy_tfidf_sw)
print("F-score: ", metrics.f1_score(corpus_test.target, predict_tfidf_sw, average='macro'))
````

````python
#Task 3.2c cont: MNB Classifier for counter-vectorized data

#instantiate classifier
mnb_clf = MultinomialNB(alpha = best_alpha_counter)

#fit (i.e. train) model to tfidf-vectorized data 
mnb_clf.fit(counter_train_sw, corpus_train.target)

#use model on test data
predict_counter_sw = mnb_clf.predict(counter_test_sw)

#calc accuracy and F-score
accuracy_counter_sw = metrics.accuracy_score(corpus_test.target, predict_counter_sw)
print("Counter-vectorized Data (no stop words) with MNB and best alpha")
print("Accuracy: ", accuracy_counter_sw)
print("F-score: ", metrics.f1_score(corpus_test.target, predict_counter_sw, average='macro'))
````
